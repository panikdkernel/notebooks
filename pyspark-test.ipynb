```
file_location = "/FileStore/tables/AB_NYC_2019.csv"

##loading the dataset
df = spark.read.format("CSV").option("inferSchema",True).option("header",True).load(file_location)
```

```
##displaying the data
display(df.take(2))
```

```
df.write.save('/FileStore/parquet/ab_nyc_2019',format='parquet')
```

```
##creating tempview for running sql
df.createOrReplaceTempView("ab_nyc")
```

```
df.columns
```

```
display(spark.sql("""
  select neighbourhood_group,count(host_name) as no_of_listings
  from ab_nyc
  group by neighbourhood_group
"""))
```
